```python
import pandas as pd
pd.set_option('display.max_rows', 5)
```

```python
movies = pd.read_csv('../data/imdb_1000.csv')
movies
```

```python
# select intersection (don't forget bracket)
print movies[(movies.duration >=200) & (movies.genre == 'Drama')]

# select union
movies[(movies.duration >=200) | (movies.genre == 'Drama')]
```

```python
# display: pretty print
from IPython.display import display
display(movies[(movies.genre == 'Crime') | (movies.genre == 'Drama') | (movies.genre == 'Action')])

# isin: contain the latter or not 
movies[movies.genre.isin(['Crime', 'Drama'])]
```
