### remove - inside "2007-08-31"

```python
dateTemp = map(lambda x: x[0:4]+x[5:7]+x[8:10], data['endDate'].values.tolist())   
datesNum = []
for date in dateTemp:
    datesNum.append(int(date))

datesNum[0:4]
```

### Try store 3 dates dataset into source dictionary

```python
sources = {}
index = 0
dates[0:5]
initialPrices = data[data.endDate == dates[0]].closePrice
len(initialPrices)
len(data[data.endDate == dates[4]].closePrice)
```

```python
for date in dates[0:5]: 
    temp = data[data.endDate == date].copy()
    temp['priceRate'] = temp.closePrice/initialPrices
    sources[datesNum[index]] = temp
    index = index + 1
    temp = None
```
