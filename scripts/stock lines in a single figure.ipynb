```python
BYD = pd.read_csv('../data/byd.csv', parse_dates=['tradeDate'])
SZZS = pd.read_csv('../data/szzs.csv', parse_dates=['tradeDate'])
MDJT = pd.read_csv('../data/mdjt.csv', parse_dates=['tradeDate'])

### Make sure date is parsed

BYD.dtypes
```

```python
display(BYD.columns)
display(SZZS.columns)
display(MDJT.columns)
```

```python
list(BYD.columns[0:2])
BYD.drop(list(BYD.columns[0:2]), inplace=True, axis=1)

list(SZZS.columns[0:3])
SZZS.drop(list(SZZS.columns[0:2]), inplace=True, axis=1)

list(MDJT.columns[0:3])
MDJT.drop(list(MDJT.columns[0:2]), inplace=True, axis=1)
```

```python
BYD.columns = ['Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Value', 'Rage']
SZZS.columns = ['Date', 'Open', 'Low', 'High', 'Close', 'Volume']
MDJT.columns = ['Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Value', 'Rate']
```

```python
### Add priceMultiple column

BYD['priceMult'] = BYD.Close/BYD.loc[0,'Close']
SZZS['priceMult'] = SZZS.Close/SZZS.loc[0,'Close']
MDJT['priceMult'] = MDJT.Close/MDJT.loc[0,'Close']

display(BYD)#.head()
display(SZZS)
display(MDJT)
```

```python
# AAPL = pd.read_csv(
#     "http://ichart.yahoo.com/table.csv?s=AAPL&a=0&b=1&c=2000&d=0&e=1&f=2015",
#     parse_dates=['Date'])
# MSFT = pd.read_csv(
#     "http://ichart.yahoo.com/table.csv?s=MSFT&a=0&b=1&c=2000&d=0&e=1&f=2015",
#     parse_dates=['Date'])
# IBM = pd.read_csv(
#     "http://ichart.yahoo.com/table.csv?s=IBM&a=0&b=1&c=2000&d=0&e=1&f=2015",
#     parse_dates=['Date'])

def stock_chart():
    p = figure(x_axis_type="datetime", width=800, height=500)

    p.line(BYD['Date'], BYD['priceMult'], color='#A6CEE3', legend='BYD')
    p.line(SZZS['Date'], SZZS['priceMult'], color='#33A02C', legend='SZZS')
    p.line(MDJT['Date'], MDJT['priceMult'], color='#FB9A99', legend='MDJT')

    p.title.text = "Stock Closing Prices"
    p.grid.grid_line_alpha=0.3
    p.xaxis.axis_label = 'Date'
    p.yaxis.axis_label = 'Price'
    p.legend.location = "top_left"
    return p
```

Embed functions all look for a Plot object. Common plot objects are:
 * `Plot` from models
 * `figure` from plotting
 * `Chart` from charts
 * `row`, `column` from layouts

```python
# Displaying in the Notebook

from bokeh.io import output_notebook, show

output_notebook()
```

```python
p = stock_chart()
show(p)
```
