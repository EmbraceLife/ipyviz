```python
p = figure(x_axis_type="datetime", width=800, height=500)

p.line(byd.tradeDate, byd.closePrice, color='#A6CEE3', legend='price')
# p.line(byd.tradeDate, byd.ETOP*1000, color='#33A02C', legend='etop')
# p.line(byd.tradeDate, byd.EPS*100, color='#33A02C', legend='EPS')
# p.line(byd.tradeDate, byd.GrossIncomeRatio*100, color='#33A02C', legend='growIncomeRate')
# p.line(byd.tradeDate, byd.IntangibleAssetRatio*1000, color='#33A02C', legend='IntangibleRate')
p.line(byd.tradeDate, byd.InventoryTRate*10, color='#33A02C', legend='inventory')

p.title.text = "Stock Closing Prices"
p.grid.grid_line_alpha=0.3
p.xaxis.axis_label = 'Date'
p.yaxis.axis_label = 'Price'
p.legend.location = "top_left"
```
