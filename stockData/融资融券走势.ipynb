# 融资融券走势

```python
import pandas as pd
%matplotlib inline
```

```python
kdxf = pd.read_csv('kdxf_rzrq.csv', encoding='GBK')
```

```python
kdxf.columns
```

```python
kdxf.sort_index(axis=0 ,ascending=False, inplace=True)
kdxf['date'] = pd.to_datetime(kdxf.tradeDate)
kdxf.set_index('date', inplace=True)
kdxf.finBuyVal.plot()
```
